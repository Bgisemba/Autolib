```python
#checking for outliers
#checking for outliers
col_names=['Rate on the coffee House compared to other brands',
       'Rate on the Price Range', 'Rate on the Importance of promotions',
       'Rate on the ambience', 'Rate on the Wifi ',
       'Rate on the service proided']

fig, ax = plt.subplots(len(col_names), figsize=(15,10))

for i, col_val in enumerate(col_names):
    sns.boxplot(df[col_val], ax=ax[i])
    ax[i].set_title('Box plot - {}'.format(col_val), fontsize=10)
    ax[i].set_xlabel(col_val, fontsize=8)
plt.show()
```

No outliers in our dataset
